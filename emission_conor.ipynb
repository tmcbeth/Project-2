```python
import pandas as pd
import numpy as np
```

```python
methane_load = "resources/totalghge2.csv"
methane_data = pd.read_csv(methane_load)
```

```python
methane_df = methane_data
```

```python
methane_df
```

```python
state_abrv = methane_df["State"]
state_greenhouse = methane_df["State nameplate capacity (MW)"]
state_emission = methane_df["State annual CH4 emissions (lbs)"]
state_co2 = methane_df["State annual CO2 emissions (tons)"]
state_biomass = methane_df["State annual biomass net generation (MWh)"]
state_transportation = methane_df["State annual total combustion net generation (MWh)"]



methane_data_df = pd.DataFrame({
    
    "State": state_abrv,
    "Tons of Greenhouse Gas Emissions": state_greenhouse,
    "Methane Emission": state_emission,
    "CO2 Emissions": state_co2,
    "State Biomass Generation": state_biomass,
    "State Transportation Generation": state_transportation
    
})


methane1_df = methane_data_df[["State", "Tons of Greenhouse Gas Emissions", "Methane Emission", "CO2 Emissions", "State Biomass Generation", "State Transportation Generation"]]
methane1_df.head()
```

```python
methane2_df = methane1_df.iloc[1:]
methane2_df
```

```python
methane2_df.dtypes
```

```python
methane2_df.sort_values(by=['State'])
```

```python
methane2_df.to_csv("./resources/methane_df.csv", header=True, index=True)
```
