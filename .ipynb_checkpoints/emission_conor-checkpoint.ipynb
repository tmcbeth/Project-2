```python
import pandas as pd
import numpy as np
```

```python
methane_load = "resources/total ghgi .csv"
methane_data = pd.read_csv(methane_load)
```

```python
methane_df = methane_data
```

```python
methane_df.head()
```

```python
state_abrv = methane_df["State abbreviation"]
state_annual = methane_df["State total annual heat input (MMBtu)"]
state_total_ozone = methane_df["State total ozone season heat input (MMBtu)"]
state_net = methane_df["State annual net generation (MWh)"]
state_net_ozone = methane_df["State ozone season net generation (MWh)"]
state_input = methane_df["State annual heat input from combustion (MMBtu)"]
state_ozone = methane_df["State ozone season heat input from combustion (MMBtu)"]
state_emission = methane_df["State annual CH4 emissions (lbs)"]
state_biomass = methane_df["State annual biomass net generation (MWh)"]



methane_data_df = pd.DataFrame({
    
    "State": state_abrv,
    "Annual Heat": state_annual,
    "Annual Ozone Input": state_total_ozone,
    "Annual Heat Input from Combustion": state_input,
    "Annual Ozone Input from Combustion": state_ozone,
    "Annual Net Generation": state_net,
    "Annual Net Ozone Generation": state_net_ozone,
    "Methane Emission": state_emission,
    "Biomass Generation": state_biomass
})


methane1_df = methane_data_df[["State", "Annual Heat", "Annual Ozone Input", "Annual Net Generation", "Annual Net Ozone Generation","Annual Heat Input from Combustion", "Annual Ozone Input from Combustion", "Methane Emission", "Biomass Generation"]]
methane1_df.head()
```

```python
methane1_df.set_index("State", inplace=True)
methane1_df.head()
```

```python
methane1_df.to_csv("./resources/methane_df.csv", header=True, index=True)
```

```python
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///resources/state_emission.sqlite")

methane1_df.to_sql('state_emission', engine, if_exists='replace', index = True)
```
